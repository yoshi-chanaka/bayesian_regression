```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd

url = 'data/auto-mpg.tsv'
df = pd.read_table(url, sep='\t', header=0)
df.info()
```

```python
df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df.drop('carname', axis=1))
plt.show()
```

```python
sns.heatmap(
    df.drop('carname', axis=1).corr(),
    annot=True,
    vmin=-1.0, vmax=1.0
)
```

```python
plt.scatter(df['weight'], df['mpg'])
plt.xlabel('weight')
plt.ylabel('mpg')
plt.savefig('figures/scatter.jpg')
plt.show()
```
